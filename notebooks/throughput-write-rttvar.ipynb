```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../bench/throughput-write-rttvar/results.csv')

# Forward fill the missing values in the RTT column
df['RTT (ms)'].fillna(method='ffill', inplace=True)

# Group by RTT and calculate the average for the write throughputs
df_grouped = df.groupby('RTT (ms)').mean()

# Reset index
df_grouped.reset_index(inplace=True)

# Create a line plot with new color scheme and legend positioning
plt.figure(figsize=(10, 6))

# Plot for Direct Mounts
plt.plot(df_grouped['RTT (ms)'], df_grouped['Write Throughput for Direct Mounts (MB/s)'], label='Direct Mounts', color='green')

# Plot for Managed Mounts
plt.plot(df_grouped['RTT (ms)'], df_grouped['Write Throughput for Managed Mounts (MB/s)'], label='Managed Mounts', color='purple')

# Set title and labels for the plot
plt.title('Average Write Throughput by RTT')
plt.xlabel('RTT (ms)')
plt.ylabel('Write Throughput (MB/s)')

# Add legend to the right center
plt.legend(loc='center right')

# Show the plot
plt.show()
```
