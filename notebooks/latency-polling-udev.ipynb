```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('../bench/latency-polling-udev/results.csv')

# Convert nanoseconds to milliseconds
data['Direct Mount Initialization Time (Polling) (ms)'] = data['Direct Mount Initialization Time (Polling) (ns)'] / 1e6

# Create KDE plot with specified style
plt.figure(figsize=(10, 6))

sns.kdeplot(data['Direct Mount Initialization Time (Polling) (ms)'], color='skyblue', fill=True, label='Polling (ms)')
sns.kdeplot(data['Direct Mount Initialization Time (udev) (ns)']/1e6, color='lightgreen', fill=True, label='udev (ms)')

plt.title('Direct Mount Initialization Time Distribution (KDE)', fontsize=15)
plt.xlabel('Initialization Time (ms)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend()
plt.show()
```
